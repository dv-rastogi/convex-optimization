```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

## Convexity of objective function
- As the given function is quadratic, we check the Hessian for positive semi-definite nature to prove its convexity
- Convexity is needed so that the minima obtained via gradient descent is the global minima
- Hessian of the given *quadratic* function is trivially calculated and is the following
$$\begin{bmatrix} 2 & -1 \\ -1 & 2*gamma \end{bmatrix}$$

```python
# Checking positive semi definite hessian
def hessian(gamma):
    return np.array([[2, -1], [-1, 2*gamma]])

print("Checking positive semi definite hessian for gamma=1:", np.all(np.linalg.eigvals(hessian(1)) > 0))
```

## Implementing function and its gradient

```python
# Function
def f(x, gamma=1):
     return x[0]**2 + gamma*x[1]**2 - x[0]*x[1] - x[0] - x[1]

# Gradient of function
def gradf(x, gamma=1):
    return np.array([2*x[0] - x[1] - 1, 2*gamma*x[1] - x[0] - 1])
```

## Implementing Gradient Descent
- Default values of precision & max iterations are taken to be 1e-12 & 1e5 respectively

```python
class GradDescent:
    def __init__(self, step_size, gamma=1):
        self.step_size = step_size
        self.gamma = gamma
    
    def error(self, prev, cur):
        return abs(f(prev, self.gamma) - f(cur, self.gamma))

    def change(self, prev, cur):
        return math.sqrt((cur[0] - prev[0])**2 + (cur[1] - prev[1])**2)

    def update(self, x):
        g = gradf(x, self.gamma)
        x = x - self.step_size * g
        return x
        
    def run(self, start=[0, 0], precision=1e-12, max_iters=1e5):
        '''
        returns iterations vs error
        '''
        X_iter = []
        Y_error = []

        cur = np.array(start.copy())
        prev_step_size = math.inf
        iters = 0
    
        while prev_step_size > precision and iters < max_iters:
            
            # Update current coord
            prev = cur.copy()
            cur = self.update(cur)

            # error
            Y_error.append(self.error(prev, cur))
            X_iter.append(iters)

            iters += 1
            prev_step_size = self.change(prev, cur)

        print(f"Global minima at x={cur} for gamma={self.gamma} & step_size={self.step_size} at iter={iters}")
        print(f"Function value reached: {f(cur, self.gamma)}")
        print(f"Precision: {precision}, Start: {start}, Max iter: {max_iters}")
        return X_iter, Y_error
```

## Part 1, Subpart 1

```python
# Parameters
grad_desc_1_1 = GradDescent(0.1)  
```

```python
X_iter, Y_error = grad_desc_1_1.run()
```

## Part 1, Subpart 2
- Step sizes were to be chosen from $$\begin{bmatrix} 0 & 2 \\ \end{bmatrix}$$
- We reformat the lower bound of the range to a value slightly greater than 0, as step size of 0 implies no descent
- New step size range $$\begin{bmatrix} 0.01 & 2 \\ \end{bmatrix}$$
- We limit our max iterations to 500

```python
# Choose 20 random values for step_size, and calculating iterations for convergence
step_sizes = np.linspace(0.01, 2, num=20)
Y_iters_step_size = [GradDescent(step_size).run(max_iters=500)[0][-1] for step_size in step_sizes]
```

```python
# Plotting step size vs iterations
plt.title("Step size vs Iterations")
plt.xlabel("step size")
plt.ylabel("iterations")
plt.plot(step_sizes, Y_iters_step_size)
plt.show()
```

## Observations
- We observe that a parabolic nature in step size vs iterations
- It resonates with our understanding of constant step size 
    - If the step size is too small, the converge will be very slow
    - If the step size is too large, divergence will occur
- We observe a suitable step size of 1 with minimal iterations in ranges of 10-20
- Also, we observe no convergence for constant step size of 2 with initial point (0, 0)
